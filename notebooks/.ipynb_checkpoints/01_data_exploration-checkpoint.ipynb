```python
# 完整的数据探索分析笔记本

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# 导入自定义模块
import sys
sys.path.append('./')
from src.data.data_loader import DataLoader
from src.data.preprocessor import DataPreprocessor
from src.features.feature_engineer import FeatureEngineer
from src.visualization.plot_utils import VisualizationUtils

# 1. 初始化
print("=== 粤港澳大湾区数据要素流动分析 ===")
print("版本: 2.0.0")
print("作者: 数据分析团队\n")

# 2. 加载数据
print("=== 数据加载 ===")
loader = DataLoader()

# 加载OD数据
od_data = loader.load_all_od_data()
print(f"OD数据: {len(od_data)}条记录")
print(f"时间范围: {od_data['年份'].min()} - {od_data['年份'].max()}")
print(f"城市数量: {od_data['source_city'].nunique()}")

# 加载面板数据
panel_data = loader.load_panel_data()
print(f"\n面板数据: {len(panel_data)}条记录")
print(f"时间范围: {panel_data['年份'].min()} - {panel_data['年份'].max()}")
print(f"变量数量: {len(panel_data.columns)}")

# 3. 数据概览
print("\n=== 数据概览 ===")

# OD数据概览
print("\nOD数据概览:")
print(od_data.head())
print("\nOD数据描述统计:")
print(od_data['value'].describe())

# 面板数据概览
print("\n面板数据概览:")
print(panel_data.head())
print("\n面板数据列名:")
print(panel_data.columns.tolist()[:20])  # 显示前20列

# 4. 数据预处理
print("\n=== 数据预处理 ===")
preprocessor = DataPreprocessor(method='mean', scale_method='standard')

# 准备分析数据
analysis_cols = [
    '跨境数据传输总量_TB',
    '数据中心数量',
    '互联网国际出口带宽_Gbps',
    'GDP_亿元',
    '数字经济核心产业增加值_亿元',
    '研发经费投入_亿元',
    '发明专利授权量',
    '5G基站数量',
    '算力规模_PFLOPS'
]

# 提取2023年数据进行分析
data_2023 = panel_data[panel_data['年份'] == 2023][['城市'] + analysis_cols].copy()

print(f"2023年数据形状: {data_2023.shape}")

# 处理缺失值
data_clean = preprocessor.handle_missing_values(data_2023[analysis_cols])

# 检测异常值
print("\n异常值检测:")
outliers_report = preprocessor.detect_outliers(data_clean)

# 处理异常值
data_processed = preprocessor.handle_outliers(data_clean)

# 标准化数据
data_scaled = preprocessor.scale_data(data_processed)

# 添加到DataFrame
for i, col in enumerate(analysis_cols):
    if i < len(data_scaled.columns):
        data_2023[f'{col}_scaled'] = data_scaled.iloc[:, i]

# 5. 探索性分析
print("\n=== 探索性分析 ===")
viz = VisualizationUtils()

# 5.1 描述性统计
print("\n1. 描述性统计:")
desc_stats = data_2023[analysis_cols].describe().T
desc_stats['cv'] = desc_stats['std'] / desc_stats['mean']  # 变异系数
desc_stats['skew'] = data_2023[analysis_cols].skew()
desc_stats['kurtosis'] = data_2023[analysis_cols].kurtosis()

print(desc_stats[['mean', 'std', 'cv', 'skew', 'kurtosis']])

# 5.2 可视化分析
print("\n2. 可视化分析:")

# 时间序列图（使用面板数据）
print("绘制时间序列图...")
viz.plot_time_series(
    panel_data[panel_data['城市'] == '深圳'],
    value_col='GDP_亿元',
    group_col='城市',
    title='深圳市GDP时间趋势',
    save_path='../outputs/figures/time_series_gdp.png'
)

# 空间分布图
print("绘制空间分布图...")
viz.plot_spatial_distribution(
    panel_data,
    value_col='跨境数据传输总量_TB',
    year=2023,
    title='2023年跨境数据传输总量分布',
    save_path='../outputs/figures/spatial_distribution.png'
)

# 相关性矩阵
print("绘制相关性矩阵...")
viz.plot_correlation_matrix(
    data_2023,
    columns=analysis_cols[:6],  # 前6个变量
    title='关键变量相关性矩阵',
    save_path='../outputs/figures/correlation_matrix.png'
)

# 箱线图比较
print("绘制箱线图比较...")
viz.plot_boxplot_comparison(
    data_2023,
    value_col='GDP_亿元',
    group_col='城市',
    title='各城市GDP比较',
    save_path='../outputs/figures/boxplot_gdp.png'
)

# 分布对比
print("绘制分布对比图...")
viz.plot_distribution_comparison(
    data_2023,
    value_cols=analysis_cols[:4],
    title='关键变量分布对比',
    save_path='../outputs/figures/distribution_comparison.png'
)

# 6. 不平等性分析
print("\n=== 不平等性分析 ===")

def calculate_gini(values):
    """计算基尼系数"""
    values = np.array(values)
    values = values.flatten()
    if np.any(values < 0):
        values = values - values.min()
    values = np.sort(values)
    n = len(values)
    index = np.arange(1, n + 1)
    gini = (np.sum((2 * index - n - 1) * values)) / (n * np.sum(values))
    return gini

def calculate_theil(values):
    """计算泰尔指数"""
    values = np.array(values)
    values = values.flatten()
    n = len(values)
    mean = np.mean(values)
    theil = (1/n) * np.sum((values/mean) * np.log(values/mean))
    return theil

# 计算各指标的不平等性
inequality_results = []
for col in analysis_cols:
    if col in data_2023.columns:
        values = data_2023[col].dropna()
        if len(values) > 0:
            gini = calculate_gini(values)
            theil = calculate_theil(values)
            inequality_results.append({
                '指标': col,
                '基尼系数': gini,
                '泰尔指数': theil
            })

inequality_df = pd.DataFrame(inequality_results)
print("\n不平等性分析结果:")
print(inequality_df.sort_values('基尼系数', ascending=False))

# 绘制洛伦兹曲线（以GDP为例）
print("\n绘制洛伦兹曲线...")
viz.plot_lorenz_curve(
    data_2023['GDP_亿元'].values,
    data_2023['城市'].values,
    title='大湾区各城市GDP洛伦兹曲线',
    save_path='../outputs/figures/lorenz_curve_gdp.png'
)

# 7. 网络分析（基于OD数据）
print("\n=== 网络分析 ===")

# 计算城市间数据流动强度
if len(od_data) > 0:
    # 按城市汇总
    city_flows = od_data.groupby('source_city')['value'].sum().reset_index()
    city_flows = city_flows.rename(columns={'source_city': '城市', 'value': '总流出量'})
    
    city_inflows = od_data.groupby('target_city')['value'].sum().reset_index()
    city_inflows = city_inflows.rename(columns={'target_city': '城市', 'value': '总流入量'})
    
    # 合并
    city_network = pd.merge(city_flows, city_inflows, on='城市', how='outer').fillna(0)
    city_network['净流量'] = city_network['总流出量'] - city_network['总流入量']
    city_network['总流量'] = city_network['总流出量'] + city_network['总流入量']
    
    print("\n城市网络流量统计:")
    print(city_network.sort_values('总流量', ascending=False))
    
    # 可视化网络流量
    viz.plot_time_series(
        od_data.groupby('年份')['value'].sum().reset_index(),
        value_col='value',
        title='大湾区总数据流量时间趋势',
        save_path='../outputs/figures/total_data_flow_trend.png'
    )

# 8. 综合仪表板
print("\n=== 创建综合仪表板 ===")

# 使用GDP作为主要指标
viz.create_dashboard(
    panel_data,
    main_metric='GDP_亿元',
    save_path='../outputs/figures/gdp_dashboard.png'
)

# 9. 保存结果
print("\n=== 保存结果 ===")

# 保存处理后的数据
data_2023.to_csv('../outputs/tables/data_2023_processed.csv', index=False)
print("处理后的数据已保存至: ../outputs/tables/data_2023_processed.csv")

# 保存不平等性分析结果
inequality_df.to_csv('../outputs/tables/inequality_analysis.csv', index=False)
print("不平等性分析结果已保存至: ../outputs/tables/inequality_analysis.csv")

# 保存网络流量统计
if 'city_network' in locals():
    city_network.to_csv('../outputs/tables/city_network_flows.csv', index=False)
    print("网络流量统计已保存至: ../outputs/tables/city_network_flows.csv")

print("\n=== 探索性数据分析完成 ===")
print(f"生成图表数量: 8张")
print(f"生成表格数量: 3个")
print(f"所有输出已保存至 ../outputs/ 目录")
```

```python
import os
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# === 关键修复：设置项目根目录 ===
# 设置项目根目录
PROJECT_ROOT = r"C:\Users\admin\粤港澳数据要素流动实验"

# 设置工作目录
os.chdir(PROJECT_ROOT)
sys.path.append(PROJECT_ROOT)

# 检查关键文件是否存在
print("检查项目结构...")
config_path = os.path.join(PROJECT_ROOT, "config", "data_config.json")
print(f"配置文件: {config_path} (存在: {os.path.exists(config_path)})")

data_dir = os.path.join(PROJECT_ROOT, "data", "raw")
print(f"数据目录: {data_dir} (存在: {os.path.exists(data_dir)})")

# 如果配置文件不存在，创建它
if not os.path.exists(config_path):
    print("创建配置文件...")
    import json
    os.makedirs(os.path.dirname(config_path), exist_ok=True)
    
    config = {
        "city_mapping": {
            "香港": 1, "澳门": 2, "广州": 3, "深圳": 4, "珠海": 5,
            "佛山": 6, "惠州": 7, "东莞": 8, "中山": 9, "江门": 10, "肇庆": 11
        },
        "analysis_vars": {
            "data_flow": ["跨境数据传输总量_TB", "数据中心数量", "互联网国际出口带宽_Gbps"],
            "economic": ["GDP_亿元", "数字经济核心产业增加值_亿元"],
            "innovation": ["研发经费投入_亿元", "发明专利授权量"],
            "infrastructure": ["5G基站数量", "算力规模_PFLOPS"]
        }
    }
    
    with open(config_path, 'w', encoding='utf-8') as f:
        json.dump(config, f, ensure_ascii=False, indent=2)
    print("配置文件已创建")

# 如果数据文件不存在，创建示例数据
if not os.path.exists(data_dir) or len(os.listdir(data_dir)) == 0:
    print("创建示例数据...")
    os.makedirs(data_dir, exist_ok=True)
    
    # 创建示例面板数据
    cities = ['香港', '澳门', '广州', '深圳', '珠海', '佛山', '惠州', '东莞', '中山', '江门', '肇庆']
    years = [2019, 2020, 2021, 2022, 2023]
    
    panel_data_list = []
    for year in years:
        for i, city in enumerate(cities):
            base = 1.08 ** (year - 2019)
            city_factor = 1 + i * 0.1
            
            panel_data_list.append({
                '城市': city,
                '年份': year,
                'GDP_亿元': round(1000 * city_factor * base, 1),
                '跨境数据传输总量_TB': round(50 * city_factor * base, 1),
                '数据中心数量': int(10 * city_factor * base),
                '研发经费投入_亿元': round(80 * city_factor * base, 1),
                '发明专利授权量': int(800 * city_factor * base),
                '5G基站数量': int(4000 * city_factor * base),
                '互联网国际出口带宽_Gbps': round(200 * city_factor * base, 1),
                '数字经济核心产业增加值_亿元': round(300 * city_factor * base, 1),
                '算力规模_PFLOPS': round(50 * city_factor * base, 1)
            })
    
    panel_df = pd.DataFrame(panel_data_list)
    panel_df.to_csv(os.path.join(data_dir, "main_data_advanced.csv"), index=False, encoding='utf-8')
    print("示例面板数据已创建")
    
    # 创建示例OD数据
    od_data_list = []
    for year in years:
        year_multiplier = 1.1 ** (year - 2019)
        for i, source in enumerate(cities):
            for j, target in enumerate(cities):
                if source != target:
                    distance_factor = 1.0 / (1 + abs(i - j) * 0.2)
                    od_data_list.append({
                        '起点城市': source,
                        '终点城市': target,
                        '数据传输量_TB': round(10 * distance_factor * (i + 1) * year_multiplier, 2),
                        '年份': year
                    })
    
    od_df = pd.DataFrame(od_data_list)
    od_df.to_csv(os.path.join(data_dir, "od_matrix.csv"), index=False, encoding='utf-8')
    print("示例OD数据已创建")

print("\n" + "="*50)
print("项目结构已准备好")
print("="*50)
# 2. 加载数据
print("=== 数据加载 ===")

# 导入模块
from src.data.data_loader import DataLoader

# 使用绝对路径初始化DataLoader
config_path = os.path.join(PROJECT_ROOT, "config", "data_config.json")
loader = DataLoader(config_path=config_path)  # 关键：传递绝对路径

# 现在继续你的代码...
# 加载OD数据
od_data = loader.load_all_od_data()

if od_data.empty:
    print("警告: OD数据为空，可能文件不存在或格式不正确")
    # 创建示例OD数据
    print("使用示例OD数据...")
    od_data = create_sample_od_data()
else:
    print(f"OD数据: {len(od_data)}条记录")
    print(f"时间范围: {od_data['年份'].min()} - {od_data['年份'].max()}")
    print(f"城市数量: {od_data['source_city'].nunique()}")

# 加载面板数据
panel_data = loader.load_panel_data()

if panel_data.empty:
    print("警告: 面板数据为空，可能文件不存在或格式不正确")
    # 创建示例面板数据
    print("使用示例面板数据...")
    panel_data = create_sample_panel_data()
else:
    print(f"\n面板数据: {len(panel_data)}条记录")
    print(f"时间范围: {panel_data['年份'].min()} - {panel_data['年份'].max()}")
    print(f"变量数量: {len(panel_data.columns)}")
    
# 3. 数据概览
print("\n=== 数据概览 ===")

# OD数据概览
print("\nOD数据概览:")
print(od_data.head())
print("\nOD数据描述统计:")
print(od_data['value'].describe())

# 面板数据概览
print("\n面板数据概览:")
print(panel_data.head())
print("\n面板数据列名:")
print(panel_data.columns.tolist()[:20])  # 显示前20列

# 4. 数据预处理
print("\n=== 数据预处理 ===")
preprocessor = DataPreprocessor(method='mean', scale_method='standard')

# 准备分析数据
analysis_cols = [
    '跨境数据传输总量_TB',
    '数据中心数量',
    '互联网国际出口带宽_Gbps',
    'GDP_亿元',
    '数字经济核心产业增加值_亿元',
    '研发经费投入_亿元',
    '发明专利授权量',
    '5G基站数量',
    '算力规模_PFLOPS'
]

# 提取2023年数据进行分析
data_2023 = panel_data[panel_data['年份'] == 2023][['城市'] + analysis_cols].copy()

print(f"2023年数据形状: {data_2023.shape}")

# 处理缺失值
data_clean = preprocessor.handle_missing_values(data_2023[analysis_cols])

# 检测异常值
print("\n异常值检测:")
outliers_report = preprocessor.detect_outliers(data_clean)

# 处理异常值
data_processed = preprocessor.handle_outliers(data_clean)

# 标准化数据
data_scaled = preprocessor.scale_data(data_processed)

# 添加到DataFrame
for i, col in enumerate(analysis_cols):
    if i < len(data_scaled.columns):
        data_2023[f'{col}_scaled'] = data_scaled.iloc[:, i]

# 5. 探索性分析
print("\n=== 探索性分析 ===")
viz = VisualizationUtils()

# 5.1 描述性统计
print("\n1. 描述性统计:")
desc_stats = data_2023[analysis_cols].describe().T
desc_stats['cv'] = desc_stats['std'] / desc_stats['mean']  # 变异系数
desc_stats['skew'] = data_2023[analysis_cols].skew()
desc_stats['kurtosis'] = data_2023[analysis_cols].kurtosis()

print(desc_stats[['mean', 'std', 'cv', 'skew', 'kurtosis']])

# 5.2 可视化分析
print("\n2. 可视化分析:")

# 时间序列图（使用面板数据）
print("绘制时间序列图...")
viz.plot_time_series(
    panel_data[panel_data['城市'] == '深圳'],
    value_col='GDP_亿元',
    group_col='城市',
    title='深圳市GDP时间趋势',
    save_path='../outputs/figures/time_series_gdp.png'
)

# 空间分布图
print("绘制空间分布图...")
viz.plot_spatial_distribution(
    panel_data,
    value_col='跨境数据传输总量_TB',
    year=2023,
    title='2023年跨境数据传输总量分布',
    save_path='../outputs/figures/spatial_distribution.png'
)

# 相关性矩阵
print("绘制相关性矩阵...")
viz.plot_correlation_matrix(
    data_2023,
    columns=analysis_cols[:6],  # 前6个变量
    title='关键变量相关性矩阵',
    save_path='../outputs/figures/correlation_matrix.png'
)

# 箱线图比较
print("绘制箱线图比较...")
viz.plot_boxplot_comparison(
    data_2023,
    value_col='GDP_亿元',
    group_col='城市',
    title='各城市GDP比较',
    save_path='../outputs/figures/boxplot_gdp.png'
)

# 分布对比
print("绘制分布对比图...")
viz.plot_distribution_comparison(
    data_2023,
    value_cols=analysis_cols[:4],
    title='关键变量分布对比',
    save_path='../outputs/figures/distribution_comparison.png'
)

# 6. 不平等性分析
print("\n=== 不平等性分析 ===")

def calculate_gini(values):
    """计算基尼系数"""
    values = np.array(values)
    values = values.flatten()
    if np.any(values < 0):
        values = values - values.min()
    values = np.sort(values)
    n = len(values)
    index = np.arange(1, n + 1)
    gini = (np.sum((2 * index - n - 1) * values)) / (n * np.sum(values))
    return gini

def calculate_theil(values):
    """计算泰尔指数"""
    values = np.array(values)
    values = values.flatten()
    n = len(values)
    mean = np.mean(values)
    theil = (1/n) * np.sum((values/mean) * np.log(values/mean))
    return theil

# 计算各指标的不平等性
inequality_results = []
for col in analysis_cols:
    if col in data_2023.columns:
        values = data_2023[col].dropna()
        if len(values) > 0:
            gini = calculate_gini(values)
            theil = calculate_theil(values)
            inequality_results.append({
                '指标': col,
                '基尼系数': gini,
                '泰尔指数': theil
            })

inequality_df = pd.DataFrame(inequality_results)
print("\n不平等性分析结果:")
print(inequality_df.sort_values('基尼系数', ascending=False))

# 绘制洛伦兹曲线（以GDP为例）
print("\n绘制洛伦兹曲线...")
viz.plot_lorenz_curve(
    data_2023['GDP_亿元'].values,
    data_2023['城市'].values,
    title='大湾区各城市GDP洛伦兹曲线',
    save_path='../outputs/figures/lorenz_curve_gdp.png'
)

# 7. 网络分析（基于OD数据）
print("\n=== 网络分析 ===")

# 计算城市间数据流动强度
if len(od_data) > 0:
    # 按城市汇总
    city_flows = od_data.groupby('source_city')['value'].sum().reset_index()
    city_flows = city_flows.rename(columns={'source_city': '城市', 'value': '总流出量'})
    
    city_inflows = od_data.groupby('target_city')['value'].sum().reset_index()
    city_inflows = city_inflows.rename(columns={'target_city': '城市', 'value': '总流入量'})
    
    # 合并
    city_network = pd.merge(city_flows, city_inflows, on='城市', how='outer').fillna(0)
    city_network['净流量'] = city_network['总流出量'] - city_network['总流入量']
    city_network['总流量'] = city_network['总流出量'] + city_network['总流入量']
    
    print("\n城市网络流量统计:")
    print(city_network.sort_values('总流量', ascending=False))
    
    # 可视化网络流量
    viz.plot_time_series(
        od_data.groupby('年份')['value'].sum().reset_index(),
        value_col='value',
        title='大湾区总数据流量时间趋势',
        save_path='../outputs/figures/total_data_flow_trend.png'
    )

# 8. 综合仪表板
print("\n=== 创建综合仪表板 ===")

# 使用GDP作为主要指标
viz.create_dashboard(
    panel_data,
    main_metric='GDP_亿元',
    save_path='../outputs/figures/gdp_dashboard.png'
)

# 9. 保存结果
print("\n=== 保存结果 ===")

# 保存处理后的数据
data_2023.to_csv('../outputs/tables/data_2023_processed.csv', index=False)
print("处理后的数据已保存至: ../outputs/tables/data_2023_processed.csv")

# 保存不平等性分析结果
inequality_df.to_csv('../outputs/tables/inequality_analysis.csv', index=False)
print("不平等性分析结果已保存至: ../outputs/tables/inequality_analysis.csv")

# 保存网络流量统计
if 'city_network' in locals():
    city_network.to_csv('../outputs/tables/city_network_flows.csv', index=False)
    print("网络流量统计已保存至: ../outputs/tables/city_network_flows.csv")

print("\n=== 探索性数据分析完成 ===")
print(f"生成图表数量: 8张")
print(f"生成表格数量: 3个")
print(f"所有输出已保存至 ../outputs/ 目录")
```

```python
# 完整的数据探索分析笔记本 - 修复版

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# 导入自定义模块
import sys
import os

# 设置项目根目录
PROJECT_ROOT = r"C:\Users\admin\粤港澳数据要素流动实验"
os.chdir(PROJECT_ROOT)
sys.path.append(PROJECT_ROOT)

from src.data.data_loader import DataLoader
from src.data.preprocessor import DataPreprocessor

# 1. 初始化
print("=== 粤港澳大湾区数据要素流动分析 ===")
print("版本: 2.0.0")
print("作者: 数据分析团队\n")

# 2. 加载数据
print("=== 数据加载 ===")

# 使用绝对路径初始化DataLoader
config_path = os.path.join(PROJECT_ROOT, "config", "data_config.json")
loader = DataLoader(config_path=config_path)

# 加载面板数据
panel_data = loader.load_panel_data()

if panel_data.empty:
    print("错误: 无法加载面板数据")
    # 创建示例数据
    print("创建示例数据...")
    cities = ['香港', '澳门', '广州', '深圳', '珠海', '佛山', '惠州', '东莞', '中山', '江门', '肇庆']
    years = [2019, 2020, 2021, 2022, 2023]
    
    data = []
    for year in years:
        for i, city in enumerate(cities):
            base = 1.08 ** (year - 2019)
            city_factor = 1 + i * 0.1
            data.append({
                '城市': city,
                '年份': year,
                'GDP_亿元': round(1000 * city_factor * base, 1),
                '跨境数据传输总量_TB': round(50 * city_factor * base, 1),
                '数据中心数量': int(10 * city_factor * base),
                '研发经费投入_亿元': round(80 * city_factor * base, 1),
                '发明专利授权量': int(800 * city_factor * base),
                '5G基站数量': int(4000 * city_factor * base),
                '互联网国际出口带宽_Gbps': round(200 * city_factor * base, 1),
                '数字经济核心产业增加值_亿元': round(300 * city_factor * base, 1),
                '算力规模_PFLOPS': round(50 * city_factor * base, 1),
                '智慧城市发展指数': round(60 + i*2 + (year-2019)*3, 1)
            })
    
    panel_data = pd.DataFrame(data)
    print(f"创建示例数据: {panel_data.shape}")

print(f"面板数据: {len(panel_data)}条记录")
print(f"时间范围: {panel_data['年份'].min()} - {panel_data['年份'].max()}")
print(f"变量数量: {len(panel_data.columns)}")

# 3. 数据概览
print("\n=== 数据概览 ===")

print("\n面板数据前5行:")
print(panel_data.head())

print("\n面板数据列名:")
print(panel_data.columns.tolist())

# 4. 数据预处理
print("\n=== 数据预处理 ===")
preprocessor = DataPreprocessor(method='mean', scale_method='standard')

# 准备分析数据
analysis_cols = []
possible_cols = [
    '跨境数据传输总量_TB',
    '数据中心数量',
    '互联网国际出口带宽_Gbps',
    'GDP_亿元',
    '数字经济核心产业增加值_亿元',
    '研发经费投入_亿元',
    '发明专利授权量',
    '5G基站数量',
    '算力规模_PFLOPS'
]

for col in possible_cols:
    if col in panel_data.columns:
        analysis_cols.append(col)

print(f"可用分析变量: {len(analysis_cols)}个")
print(analysis_cols)

# 提取2023年数据进行分析
if 2023 in panel_data['年份'].values:
    data_2023 = panel_data[panel_data['年份'] == 2023][['城市'] + analysis_cols].copy()
else:
    # 使用最新年份
    latest_year = panel_data['年份'].max()
    data_2023 = panel_data[panel_data['年份'] == latest_year][['城市'] + analysis_cols].copy()
    print(f"注意: 使用{latest_year}年数据代替2023年")

print(f"分析数据形状: {data_2023.shape}")

# 5. 探索性分析
print("\n=== 探索性分析 ===")

# 5.1 描述性统计
print("\n1. 描述性统计:")
desc_stats = data_2023[analysis_cols].describe().T
desc_stats['cv'] = desc_stats['std'] / desc_stats['mean']  # 变异系数
desc_stats['skew'] = data_2023[analysis_cols].skew()
desc_stats['kurtosis'] = data_2023[analysis_cols].kurtosis()

print(desc_stats[['mean', 'std', 'cv', 'skew', 'kurtosis']])

# 5.2 可视化分析
print("\n2. 可视化分析:")

# 创建输出目录
output_dir = os.path.join(PROJECT_ROOT, "outputs", "figures")
os.makedirs(output_dir, exist_ok=True)

# 1) 时间序列图（使用面板数据）
print("绘制时间序列图...")
if '深圳' in panel_data['城市'].values and 'GDP_亿元' in panel_data.columns:
    shenzhen_data = panel_data[panel_data['城市'] == '深圳'].sort_values('年份')
    
    plt.figure(figsize=(12, 6))
    plt.plot(shenzhen_data['年份'], shenzhen_data['GDP_亿元'], 'bo-', linewidth=2, markersize=8)
    plt.xlabel('年份')
    plt.ylabel('GDP (亿元)')
    plt.title('深圳市GDP时间趋势')
    plt.grid(True, alpha=0.3)
    
    # 添加数据标签
    for x, y in zip(shenzhen_data['年份'], shenzhen_data['GDP_亿元']):
        plt.text(x, y, f'{y:.1f}', ha='center', va='bottom')
    
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, "time_series_gdp.png"), dpi=300)
    plt.show()

# 2) 空间分布图（2023年各城市比较）
print("绘制空间分布图...")
plt.figure(figsize=(14, 8))

if '跨境数据传输总量_TB' in data_2023.columns:
    sorted_data = data_2023.sort_values('跨境数据传输总量_TB', ascending=False)
    
    colors = plt.cm.Set3(np.linspace(0, 1, len(sorted_data)))
    bars = plt.bar(range(len(sorted_data)), sorted_data['跨境数据传输总量_TB'], color=colors)
    
    plt.xticks(range(len(sorted_data)), sorted_data['城市'], rotation=45, ha='right')
    plt.xlabel('城市')
    plt.ylabel('跨境数据传输总量 (TB)')
    plt.title('2023年各城市跨境数据传输总量分布')
    plt.grid(True, alpha=0.3, axis='y')
    
    # 添加数值标签
    for i, (bar, value) in enumerate(zip(bars, sorted_data['跨境数据传输总量_TB'])):
        plt.text(i, value + max(sorted_data['跨境数据传输总量_TB'])*0.01, 
                f'{value:.1f}', ha='center', va='bottom', fontsize=9)
    
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, "spatial_distribution.png"), dpi=300)
    plt.show()

# 3) 相关性矩阵
print("绘制相关性矩阵...")
if len(analysis_cols) >= 2:
    # 计算相关性矩阵
    corr_matrix = data_2023[analysis_cols[:min(6, len(analysis_cols))]].corr()
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, 
                square=True, fmt='.2f', cbar_kws={"shrink": 0.8})
    plt.title('关键变量相关性矩阵')
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, "correlation_matrix.png"), dpi=300)
    plt.show()

# 4) 箱线图比较
print("绘制箱线图比较...")
if 'GDP_亿元' in data_2023.columns:
    plt.figure(figsize=(12, 6))
    
    # 按GDP排序
    sorted_cities = data_2023.sort_values('GDP_亿元')['城市']
    
    # 使用boxplot
    box_data = [panel_data[panel_data['城市'] == city]['GDP_亿元'].values 
                for city in sorted_cities if city in panel_data['城市'].values]
    
    plt.boxplot(box_data, labels=sorted_cities)
    plt.xticks(rotation=45, ha='right')
    plt.xlabel('城市')
    plt.ylabel('GDP (亿元)')
    plt.title('各城市GDP分布比较')
    plt.grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, "boxplot_gdp.png"), dpi=300)
    plt.show()

# 5) 分布对比
print("绘制分布对比图...")
if len(analysis_cols) >= 4:
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    axes = axes.flatten()
    
    for i, col in enumerate(analysis_cols[:4]):
        if col in data_2023.columns:
            axes[i].hist(data_2023[col].dropna(), bins=15, alpha=0.7, color='steelblue', edgecolor='black')
            axes[i].set_xlabel(col)
            axes[i].set_ylabel('频数')
            axes[i].set_title(f'{col}分布')
            axes[i].grid(True, alpha=0.3)
    
    plt.suptitle('关键变量分布对比', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, "distribution_comparison.png"), dpi=300)
    plt.show()

# 6. 不平等性分析
print("\n=== 不平等性分析 ===")

def calculate_gini(values):
    """计算基尼系数"""
    values = np.array(values)
    values = values.flatten()
    if np.any(values < 0):
        values = values - values.min()
    values = values + 1e-10  # 避免除零
    values = np.sort(values)
    n = len(values)
    index = np.arange(1, n + 1)
    gini = (np.sum((2 * index - n - 1) * values)) / (n * np.sum(values))
    return gini

def calculate_theil(values):
    """计算泰尔指数"""
    values = np.array(values)
    values = values.flatten()
    values = values + 1e-10  # 避免除零
    n = len(values)
    mean = np.mean(values)
    theil = (1/n) * np.sum((values/mean) * np.log(values/mean + 1e-10))
    return theil

# 计算各指标的不平等性
inequality_results = []
for col in analysis_cols:
    if col in data_2023.columns:
        values = data_2023[col].dropna()
        if len(values) > 1:
            gini = calculate_gini(values)
            theil = calculate_theil(values)
            inequality_results.append({
                '指标': col,
                '基尼系数': gini,
                '泰尔指数': theil
            })

if inequality_results:
    inequality_df = pd.DataFrame(inequality_results)
    print("\n不平等性分析结果:")
    print(inequality_df.sort_values('基尼系数', ascending=False))
    
    # 绘制不平等性结果
    plt.figure(figsize=(12, 6))
    x = range(len(inequality_df))
    width = 0.35
    
    plt.bar([i - width/2 for i in x], inequality_df['基尼系数'], width, label='基尼系数', alpha=0.7)
    plt.bar([i + width/2 for i in x], inequality_df['泰尔指数'], width, label='泰尔指数', alpha=0.7)
    
    plt.xticks(x, inequality_df['指标'], rotation=45, ha='right')
    plt.ylabel('指数值')
    plt.title('各指标不平等性分析')
    plt.legend()
    plt.grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, "inequality_analysis.png"), dpi=300)
    plt.show()

# 7. 保存结果
print("\n=== 保存结果 ===")

# 创建tables目录
tables_dir = os.path.join(PROJECT_ROOT, "outputs", "tables")
os.makedirs(tables_dir, exist_ok=True)

# 保存处理后的数据
data_2023.to_csv(os.path.join(tables_dir, "data_2023_processed.csv"), index=False, encoding='utf-8')
print(f"处理后的数据已保存至: {os.path.join(tables_dir, 'data_2023_processed.csv')}")

# 保存不平等性分析结果
if inequality_results:
    inequality_df.to_csv(os.path.join(tables_dir, "inequality_analysis.csv"), index=False, encoding='utf-8')
    print(f"不平等性分析结果已保存至: {os.path.join(tables_dir, 'inequality_analysis.csv')}")

# 保存描述性统计
desc_stats.to_csv(os.path.join(tables_dir, "descriptive_statistics.csv"), encoding='utf-8')
print(f"描述性统计已保存至: {os.path.join(tables_dir, 'descriptive_statistics.csv')}")

print("\n" + "="*60)
print("探索性数据分析完成!")
print(f"生成图表数量: 5张")
print(f"生成表格数量: 3个")
print(f"所有输出已保存至 {os.path.join(PROJECT_ROOT, 'outputs')} 目录")
print("="*60)
```

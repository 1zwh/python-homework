```python
# PCA和因子分析笔记本

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity, calculate_kmo
import warnings
warnings.filterwarnings('ignore')

# 导入自定义模块
import sys
sys.path.append('../')
from src.data.data_loader import DataLoader
from src.data.preprocessor import DataPreprocessor
from src.models.pca_analyzer import PCAAnalyzer
from src.visualization.plot_utils import VisualizationUtils

# 1. 数据准备
print("=== PCA与因子分析 ===")

# 加载数据
loader = DataLoader()
panel_data = loader.load_panel_data()

# 选择2023年数据进行分析
data_2023 = panel_data[panel_data['年份'] == 2023].copy()

# 选择分析变量
analysis_vars = [
    '跨境数据传输总量_TB',
    '数据中心数量',
    '互联网国际出口带宽_Gbps',
    'IaaS使用量_万核时',
    '数据交易额_亿元',
    'GDP_亿元',
    '数字经济核心产业增加值_亿元',
    '研发经费投入_亿元',
    '发明专利授权量',
    '5G基站数量',
    '算力规模_PFLOPS',
    '智慧城市发展指数'
]

# 检查变量是否存在
existing_vars = [var for var in analysis_vars if var in data_2023.columns]
print(f"使用的变量: {len(existing_vars)}个")

# 提取数据
X_raw = data_2023[existing_vars].values
city_labels = data_2023['城市'].values

# 2. 数据预处理
print("\n=== 数据预处理 ===")
preprocessor = DataPreprocessor(method='mean', scale_method='standard')

# 处理缺失值
X_clean = preprocessor.handle_missing_values(pd.DataFrame(X_raw, columns=existing_vars))

# 标准化
X_scaled = preprocessor.scale_data(X_clean)

print(f"数据形状: {X_scaled.shape}")
print(f"城市数量: {len(city_labels)}")

# 3. 因子分析适用性检验
print("\n=== 因子分析适用性检验 ===")

# KMO检验
kmo_all, kmo_model = calculate_kmo(X_scaled.values)
print(f"KMO统计量: {kmo_model:.4f}")

# Bartlett球形检验
chi_square, p_value = calculate_bartlett_sphericity(X_scaled.values)
print(f"Bartlett球形检验:")
print(f"  卡方值: {chi_square:.2f}")
print(f"  P值: {p_value:.6f}")
print(f"  结论: {'适合因子分析' if p_value < 0.05 else '不适合因子分析'}")

# 4. 主成分分析
print("\n=== 主成分分析 ===")

# 创建PCA分析器
pca_analyzer = PCAAnalyzer()
pca_analyzer.fit(X_scaled.values, feature_names=existing_vars)

# 确定最优主成分数量
print("\n确定最优主成分数量...")
optimal_components = pca_analyzer.determine_optimal_components(X_scaled.values)
print("各方法建议的主成分数量:")
for method, n in optimal_components.items():
    print(f"  {method}: {n}个")

# 使用多数原则确定最终数量
final_n = max(optimal_components.values())
print(f"\n最终选择主成分数量: {final_n}")

# 重新拟合PCA
pca_final = PCAAnalyzer(n_components=final_n)
pca_final.fit(X_scaled.values, feature_names=existing_vars)

# 5. PCA结果分析
print("\n=== PCA结果分析 ===")

# 获取PCA摘要
pca_summary = pca_final.get_pca_summary()
print("\n主成分摘要:")
print(pca_summary.to_string(index=False))

# 特征值和方差贡献率
print("\n特征值和方差贡献率:")
for i, (eigenvalue, ratio) in enumerate(zip(pca_final.eigenvalues_, 
                                          pca_final.explained_variance_ratio_)):
    cum_ratio = np.sum(pca_final.explained_variance_ratio_[:i+1])
    print(f"PC{i+1}: 特征值={eigenvalue:.3f}, "
          f"贡献率={ratio*100:.1f}%, "
          f"累计={cum_ratio*100:.1f}%")

# 特征载荷矩阵
print("\n特征载荷矩阵（绝对值>0.5）:")
loadings_df = pca_final.loadings_df.copy()
# 高亮显示高载荷
loadings_display = loadings_df.applymap(
    lambda x: f"{x:.3f}" if abs(x) >= 0.5 else f"({x:.3f})"
)
print(loadings_display)

# 6. 可视化
print("\n=== 可视化 ===")
viz = VisualizationUtils()

# 绘制碎石图
print("绘制碎石图...")
pca_final.plot_scree(save_path='../outputs/figures/pca_scree_plot.png')

# 绘制双标图
print("绘制双标图...")
pca_final.plot_biplot(X_scaled.values, labels=city_labels,
                     save_path='../outputs/figures/pca_biplot.png')

# 绘制载荷热力图
print("绘制载荷热力图...")
pca_final.plot_loadings_heatmap(save_path='../outputs/figures/pca_loadings_heatmap.png')

# 7. 因子分析
print("\n=== 探索性因子分析 ===")

# 确定因子数量
print("确定最优因子数量...")
n_factors_range = range(1, min(6, len(existing_vars)//2 + 1))
results = []

for n in n_factors_range:
    fa = FactorAnalyzer(n_factors=n, rotation='varimax')
    fa.fit(X_scaled.values)
    
    # 计算RMSEA（如果可用）
    try:
        rmsea = fa.get_factor_variance()['RMSEA'][0]
    except:
        rmsea = None
    
    results.append({
        'n_factors': n,
        'rmsea': rmsea
    })

results_df = pd.DataFrame(results)
print("\n不同因子数量比较:")
print(results_df)

# 选择RMSEA最小的因子数量
if results_df['rmsea'].notna().any():
    optimal_n_factors = results_df.loc[results_df['rmsea'].idxmin(), 'n_factors']
else:
    # 如果没有RMSEA，使用特征值>1准则
    optimal_n_factors = optimal_components['eigenvalue']

print(f"\n最优因子数量: {optimal_n_factors}")

# 进行因子分析
fa_final = FactorAnalyzer(n_factors=optimal_n_factors, rotation='varimax')
fa_final.fit(X_scaled.values)

# 因子载荷矩阵
loadings_fa = fa_final.loadings_
loadings_df_fa = pd.DataFrame(
    loadings_fa,
    columns=[f'Factor{i+1}' for i in range(optimal_n_factors)],
    index=existing_vars
)

print("\n因子载荷矩阵:")
print(loadings_df_fa.round(3))

# 计算因子得分
factor_scores = fa_final.transform(X_scaled.values)
factor_scores_df = pd.DataFrame(
    factor_scores,
    columns=[f'Factor{i+1}_Score' for i in range(optimal_n_factors)],
    index=city_labels
)

# 计算综合得分
factor_variance = fa_final.get_factor_variance()
weights = factor_variance['Proportional Var'][:optimal_n_factors]
weights = weights / weights.sum()

factor_scores_df['综合得分'] = np.dot(factor_scores, weights)
factor_scores_df['排名'] = factor_scores_df['综合得分'].rank(ascending=False).astype(int)

print("\n城市综合得分排名:")
print(factor_scores_df[['综合得分', '排名']].sort_values('排名'))

# 8. 可视化因子分析结果
print("\n=== 因子分析可视化 ===")

# 绘制因子载荷热力图
fig, ax = plt.subplots(figsize=(12, 8))
sns.heatmap(loadings_df_fa, annot=True, cmap='RdBu_r', center=0,
           ax=ax, cbar_kws={"shrink": 0.8})
ax.set_title('因子载荷矩阵热力图', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('../outputs/figures/factor_loadings_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

# 绘制综合得分条形图
fig, ax = plt.subplots(figsize=(12, 8))
sorted_scores = factor_scores_df.sort_values('综合得分', ascending=False)
colors = plt.cm.Set3(np.linspace(0, 1, len(sorted_scores)))

bars = ax.barh(range(len(sorted_scores)), sorted_scores['综合得分'], color=colors)
ax.set_yticks(range(len(sorted_scores)))
ax.set_yticklabels(sorted_scores.index)
ax.set_xlabel('综合得分')
ax.set_title('大湾区城市数据要素发展水平综合得分', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, axis='x')

# 添加数值标签
for i, (bar, score) in enumerate(zip(bars, sorted_scores['综合得分'])):
    ax.text(score + 0.01, bar.get_y() + bar.get_height()/2, 
           f'{score:.3f}', va='center', fontsize=10)

plt.tight_layout()
plt.savefig('../outputs/figures/comprehensive_scores.png', dpi=300, bbox_inches='tight')
plt.show()

# 9. 稳健性检验
print("\n=== 稳健性检验 ===")

# 使用不同PCA方法比较
methods_to_test = ['standard', 'robust', 'kernel_rbf']

robustness_results = {}

for method in methods_to_test:
    print(f"\n测试方法: {method}")
    
    if method == 'standard':
        pca_test = PCAAnalyzer(n_components=final_n)
    elif method == 'robust':
        pca_test = PCAAnalyzer(n_components=final_n, robust=True)
    elif method == 'kernel_rbf':
        pca_test = PCAAnalyzer(n_components=final_n, kernel='rbf')
    
    pca_test.fit(X_scaled.values, feature_names=existing_vars)
    
    # 比较前两个主成分的载荷相关性
    if hasattr(pca_test, 'loadings_') and hasattr(pca_final, 'loadings_'):
        corr_pc1 = np.corrcoef(pca_final.loadings_[0], pca_test.loadings_[0])[0, 1]
        corr_pc2 = np.corrcoef(pca_final.loadings_[1], pca_test.loadings_[1])[0, 1]
        
        robustness_results[method] = {
            'PC1_Correlation': corr_pc1,
            'PC2_Correlation': corr_pc2
        }
        
        print(f"  与标准PCA的载荷相关性:")
        print(f"    PC1: {corr_pc1:.4f}")
        print(f"    PC2: {corr_pc2:.4f}")

print("\n稳健性检验汇总:")
for method, results in robustness_results.items():
    print(f"{method}: PC1相关性={results['PC1_Correlation']:.4f}, "
          f"PC2相关性={results['PC2_Correlation']:.4f}")

# 10. 保存结果
print("\n=== 保存结果 ===")

# 保存PCA摘要
pca_summary.to_csv('../outputs/tables/pca_summary.csv', index=False)
print("PCA摘要已保存至: ../outputs/tables/pca_summary.csv")

# 保存载荷矩阵
loadings_df.to_csv('../outputs/tables/pca_loadings.csv')
print("PCA载荷矩阵已保存至: ../outputs/tables/pca_loadings.csv")

# 保存因子载荷矩阵
loadings_df_fa.to_csv('../outputs/tables/factor_loadings.csv')
print("因子载荷矩阵已保存至: ../outputs/tables/factor_loadings.csv")

# 保存综合得分
factor_scores_df.to_csv('../outputs/tables/comprehensive_scores.csv')
print("综合得分已保存至: ../outputs/tables/comprehensive_scores.csv")

# 保存稳健性检验结果
robustness_df = pd.DataFrame(robustness_results).T
robustness_df.to_csv('../outputs/tables/robustness_test.csv')
print("稳健性检验结果已保存至: ../outputs/tables/robustness_test.csv")

print("\n=== PCA与因子分析完成 ===")
print(f"生成图表数量: 5张")
print(f"生成表格数量: 5个")
print(f"所有输出已保存至 ../outputs/ 目录")
```

```python
# PCA和因子分析笔记本 - 修复路径版

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity, calculate_kmo
import warnings
import os
import sys

warnings.filterwarnings('ignore')

# === 1. 设置路径 ===
# 设置项目根目录
PROJECT_ROOT = r"C:\Users\admin\粤港澳数据要素流动实验"
print(f"项目根目录: {PROJECT_ROOT}")

# 设置工作目录
os.chdir(PROJECT_ROOT)
sys.path.append(PROJECT_ROOT)

# 创建输出目录
outputs_figures = os.path.join(PROJECT_ROOT, "outputs", "figures")
outputs_tables = os.path.join(PROJECT_ROOT, "outputs", "tables")
os.makedirs(outputs_figures, exist_ok=True)
os.makedirs(outputs_tables, exist_ok=True)

print(f"输出目录已创建:")
print(f"  图表目录: {outputs_figures}")
print(f"  表格目录: {outputs_tables}")

# 导入自定义模块
from src.data.data_loader import DataLoader
from src.data.preprocessor import DataPreprocessor
from src.models.pca_analyzer import PCAAnalyzer

# === 2. 数据准备 ===
print("\n" + "="*60)
print("PCA与因子分析")
print("="*60)

# 加载数据
config_path = os.path.join(PROJECT_ROOT, "config", "data_config.json")
loader = DataLoader(config_path=config_path)

panel_data = loader.load_panel_data()

if panel_data.empty:
    print("错误: 无法加载面板数据")
    # 创建示例数据
    cities = ['香港', '澳门', '广州', '深圳', '珠海', '佛山', '惠州', '东莞', '中山', '江门', '肇庆']
    data = []
    for city in cities:
        for year in [2019, 2020, 2021, 2022, 2023]:
            data.append({
                '城市': city,
                '年份': year,
                '跨境数据传输总量_TB': np.random.uniform(50, 500),
                '数据中心数量': np.random.randint(5, 50),
                '互联网国际出口带宽_Gbps': np.random.uniform(100, 1000),
                'GDP_亿元': np.random.uniform(500, 5000),
                '数字经济核心产业增加值_亿元': np.random.uniform(100, 1000),
                '研发经费投入_亿元': np.random.uniform(30, 300),
                '发明专利授权量': np.random.randint(100, 5000),
                '5G基站数量': np.random.randint(1000, 10000),
                '算力规模_PFLOPS': np.random.uniform(10, 200),
                '智慧城市发展指数': np.random.uniform(50, 95)
            })
    panel_data = pd.DataFrame(data)
    print("使用生成的示例数据")

# 选择2023年数据进行分析
if 2023 in panel_data['年份'].values:
    data_2023 = panel_data[panel_data['年份'] == 2023].copy()
else:
    # 使用最新年份
    latest_year = panel_data['年份'].max()
    data_2023 = panel_data[panel_data['年份'] == latest_year].copy()
    print(f"注意: 使用{latest_year}年数据代替2023年")

print(f"分析数据年份: {data_2023['年份'].iloc[0]}")

# 选择分析变量
analysis_vars = [
    '跨境数据传输总量_TB',
    '数据中心数量',
    '互联网国际出口带宽_Gbps',
    'GDP_亿元',
    '数字经济核心产业增加值_亿元',
    '研发经费投入_亿元',
    '发明专利授权量',
    '5G基站数量',
    '算力规模_PFLOPS',
    '智慧城市发展指数'
]

# 检查变量是否存在
existing_vars = [var for var in analysis_vars if var in data_2023.columns]
print(f"\n使用的变量 ({len(existing_vars)}个):")
for var in existing_vars:
    print(f"  - {var}")

if len(existing_vars) < 3:
    print("错误: 可用变量太少，无法进行PCA分析")
    sys.exit(1)

# 提取数据
X_raw = data_2023[existing_vars].values
city_labels = data_2023['城市'].values

# === 3. 数据预处理 ===
print("\n" + "="*60)
print("数据预处理")
print("="*60)

preprocessor = DataPreprocessor(method='mean', scale_method='standard')

# 处理缺失值
X_clean = preprocessor.handle_missing_values(pd.DataFrame(X_raw, columns=existing_vars))

# 标准化
X_scaled = preprocessor.scale_data(X_clean)

print(f"数据形状: {X_scaled.shape}")
print(f"城市数量: {len(city_labels)}")

# === 4. 因子分析适用性检验 ===
print("\n" + "="*60)
print("因子分析适用性检验")
print("="*60)

try:
    # KMO检验
    kmo_all, kmo_model = calculate_kmo(X_scaled.values)
    print(f"KMO统计量: {kmo_model:.4f}")
    
    # Bartlett球形检验
    chi_square, p_value = calculate_bartlett_sphericity(X_scaled.values)
    print(f"Bartlett球形检验:")
    print(f"  卡方值: {chi_square:.2f}")
    print(f"  P值: {p_value:.6f}")
    print(f"  结论: {'适合因子分析' if p_value < 0.05 else '不适合因子分析'}")
except Exception as e:
    print(f"因子分析适用性检验出错: {e}")
    print("跳过适用性检验...")

# === 5. 主成分分析 ===
print("\n" + "="*60)
print("主成分分析")
print("="*60)

# 创建PCA分析器
pca_analyzer = PCAAnalyzer()
pca_analyzer.fit(X_scaled.values, feature_names=existing_vars)

# 确定最优主成分数量
print("\n确定最优主成分数量...")

# 使用特征值>1准则
eigenvalues = pca_analyzer.eigenvalues_
n_components = np.sum(eigenvalues > 1)
print(f"特征值>1准则: {n_components}个主成分")

# 累计方差≥80%
cumulative_variance = np.cumsum(pca_analyzer.explained_variance_ratio_)
n_variance = np.argmax(cumulative_variance >= 0.80) + 1
if n_variance == 1 and cumulative_variance[0] < 0.80:
    n_variance = len(cumulative_variance)
print(f"累计方差≥80%: {n_variance}个主成分")

# 选择主成分数量
final_n = max(2, min(n_components, n_variance, 4))
print(f"\n最终选择主成分数量: {final_n}个")

# 重新拟合PCA
pca_final = PCAAnalyzer(n_components=final_n)
pca_final.fit(X_scaled.values, feature_names=existing_vars)

# === 6. PCA结果分析 ===
print("\n" + "="*60)
print("PCA结果分析")
print("="*60)

# 获取PCA摘要
try:
    pca_summary = pca_final.get_pca_summary()
    print("\n主成分摘要:")
    print(pca_summary.to_string(index=False))
except:
    # 手动创建摘要
    pca_summary = pd.DataFrame({
        'PC': [f'PC{i+1}' for i in range(final_n)],
        'Eigenvalue': pca_final.eigenvalues_[:final_n],
        'Variance_Explained': pca_final.explained_variance_ratio_[:final_n],
        'Cumulative_Variance': np.cumsum(pca_final.explained_variance_ratio_[:final_n])
    })
    print("\n主成分摘要:")
    print(pca_summary)

# 特征值和方差贡献率
print("\n特征值和方差贡献率:")
for i, (eigenvalue, ratio) in enumerate(zip(pca_final.eigenvalues_[:final_n], 
                                          pca_final.explained_variance_ratio_[:final_n])):
    cum_ratio = np.sum(pca_final.explained_variance_ratio_[:i+1])
    print(f"PC{i+1}: 特征值={eigenvalue:.3f}, "
          f"贡献率={ratio*100:.1f}%, "
          f"累计={cum_ratio*100:.1f}%")

# === 7. 可视化 ===
print("\n" + "="*60)
print("可视化")
print("="*60)

# 绘制碎石图
print("绘制碎石图...")
plt.figure(figsize=(12, 5))
x = range(1, len(eigenvalues) + 1)
plt.subplot(1, 2, 1)
plt.plot(x, eigenvalues, 'bo-', linewidth=2, markersize=8)
plt.axhline(y=1, color='r', linestyle='--', label='特征值=1')
plt.xlabel('主成分序号')
plt.ylabel('特征值')
plt.title('PCA碎石图')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
cumulative_variance = np.cumsum(pca_analyzer.explained_variance_ratio_)
plt.bar(x, pca_analyzer.explained_variance_ratio_, alpha=0.7, label='单个贡献率')
plt.plot(x, cumulative_variance, 'ro-', linewidth=2, label='累计贡献率')
plt.axhline(y=0.80, color='g', linestyle='--', label='80%阈值')
plt.xlabel('主成分序号')
plt.ylabel('方差贡献率')
plt.title('方差贡献率')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(outputs_figures, "pca_scree_plot.png"), dpi=300, bbox_inches='tight')
plt.show()

# 绘制双标图
print("绘制双标图...")
try:
    # 转换数据
    if hasattr(pca_final, 'scaler'):
        X_scaled_for_plot = pca_final.scaler.transform(X_raw)
    else:
        X_scaled_for_plot = X_scaled.values
    
    # 获取主成分得分
    if hasattr(pca_final, 'pca'):
        scores = pca_final.pca.transform(X_scaled_for_plot)
    else:
        scores = pca_final.transform(X_scaled_for_plot)
    
    plt.figure(figsize=(10, 8))
    scatter = plt.scatter(scores[:, 0], scores[:, 1], alpha=0.7, s=100, edgecolors='k')
    
    # 添加标签
    for i, label in enumerate(city_labels):
        plt.annotate(label, (scores[i, 0], scores[i, 1]),
                    xytext=(5, 5), textcoords='offset points',
                    fontsize=10, fontweight='bold')
    
    plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    
    plt.xlabel(f'PC1 ({pca_final.explained_variance_ratio_[0]*100:.1f}%)')
    plt.ylabel(f'PC2 ({pca_final.explained_variance_ratio_[1]*100:.1f}%)')
    plt.title('PCA双标图')
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(os.path.join(outputs_figures, "pca_biplot.png"), dpi=300, bbox_inches='tight')
    plt.show()
except Exception as e:
    print(f"绘制双标图时出错: {e}")

# === 8. 因子分析 ===
print("\n" + "="*60)
print("探索性因子分析")
print("="*60)

try:
    # 确定因子数量
    n_factors_range = range(1, min(4, len(existing_vars)//2 + 1))
    results = []
    
    for n in n_factors_range:
        fa = FactorAnalyzer(n_factors=n, rotation='varimax')
        fa.fit(X_scaled.values)
        
        results.append({
            'n_factors': n,
            'eigenvalues': fa.get_eigenvalues()[0][:n].sum()
        })
    
    results_df = pd.DataFrame(results)
    print("\n不同因子数量比较:")
    print(results_df)
    
    # 选择特征值总和最大的因子数量
    optimal_n_factors = results_df.loc[results_df['eigenvalues'].idxmax(), 'n_factors']
    print(f"\n最优因子数量: {optimal_n_factors}")
    
    # 进行因子分析
    fa_final = FactorAnalyzer(n_factors=optimal_n_factors, rotation='varimax')
    fa_final.fit(X_scaled.values)
    
    # 因子载荷矩阵
    loadings_fa = fa_final.loadings_
    loadings_df_fa = pd.DataFrame(
        loadings_fa,
        columns=[f'Factor{i+1}' for i in range(optimal_n_factors)],
        index=existing_vars
    )
    
    print("\n因子载荷矩阵:")
    print(loadings_df_fa.round(3))
    
    # 计算因子得分
    factor_scores = fa_final.transform(X_scaled.values)
    factor_scores_df = pd.DataFrame(
        factor_scores,
        columns=[f'Factor{i+1}_Score' for i in range(optimal_n_factors)],
        index=city_labels
    )
    
    # 计算综合得分（简单平均）
    factor_scores_df['综合得分'] = factor_scores_df.mean(axis=1)
    factor_scores_df['排名'] = factor_scores_df['综合得分'].rank(ascending=False).astype(int)
    
    print("\n城市综合得分排名:")
    print(factor_scores_df[['综合得分', '排名']].sort_values('排名'))
    
    # 绘制因子载荷热力图
    plt.figure(figsize=(12, 8))
    sns.heatmap(loadings_df_fa, annot=True, cmap='RdBu_r', center=0,
               cbar_kws={"shrink": 0.8})
    plt.title('因子载荷矩阵热力图', fontsize=14, fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig(os.path.join(outputs_figures, "factor_loadings_heatmap.png"), dpi=300, bbox_inches='tight')
    plt.show()
    
    # 绘制综合得分条形图
    plt.figure(figsize=(12, 8))
    sorted_scores = factor_scores_df.sort_values('综合得分', ascending=False)
    colors = plt.cm.Set3(np.linspace(0, 1, len(sorted_scores)))
    
    bars = plt.barh(range(len(sorted_scores)), sorted_scores['综合得分'], color=colors)
    plt.yticks(range(len(sorted_scores)), sorted_scores.index)
    plt.xlabel('综合得分')
    plt.title('大湾区城市数据要素发展水平综合得分', fontsize=14, fontweight='bold')
    plt.grid(True, alpha=0.3, axis='x')
    
    # 添加数值标签
    for i, (bar, score) in enumerate(zip(bars, sorted_scores['综合得分'])):
        plt.text(score + 0.01, bar.get_y() + bar.get_height()/2, 
                f'{score:.3f}', va='center', fontsize=10)
    
    plt.tight_layout()
    plt.savefig(os.path.join(outputs_figures, "comprehensive_scores.png"), dpi=300, bbox_inches='tight')
    plt.show()
    
except Exception as e:
    print(f"因子分析出错: {e}")
    print("跳过因子分析...")

# === 9. 保存结果 ===
print("\n" + "="*60)
print("保存结果")
print("="*60)

# 保存PCA摘要
pca_summary.to_csv(os.path.join(outputs_tables, "pca_summary.csv"), index=False, encoding='utf-8')
print(f"PCA摘要已保存: {os.path.join(outputs_tables, 'pca_summary.csv')}")

# 保存主成分得分
pca_scores = pca_final.transform(X_scaled.values)
pca_scores_df = pd.DataFrame(
    pca_scores,
    columns=[f'PC{i+1}' for i in range(final_n)],
    index=city_labels
)
pca_scores_df.to_csv(os.path.join(outputs_tables, "pca_scores.csv"), encoding='utf-8')
print(f"PCA得分已保存: {os.path.join(outputs_tables, 'pca_scores.csv')}")

# 保存综合得分（如果进行了因子分析）
if 'factor_scores_df' in locals():
    factor_scores_df.to_csv(os.path.join(outputs_tables, "comprehensive_scores.csv"), encoding='utf-8')
    print(f"综合得分已保存: {os.path.join(outputs_tables, 'comprehensive_scores.csv')}")
    
    # 保存因子载荷矩阵
    loadings_df_fa.to_csv(os.path.join(outputs_tables, "factor_loadings.csv"), encoding='utf-8')
    print(f"因子载荷矩阵已保存: {os.path.join(outputs_tables, 'factor_loadings.csv')}")

print("\n" + "="*60)
print("PCA与因子分析完成!")
print(f"生成图表数量: 3-4张")
print(f"生成表格数量: 3-4个")
print(f"所有输出已保存至: {os.path.join(PROJECT_ROOT, 'outputs')}")
print("="*60)
```
